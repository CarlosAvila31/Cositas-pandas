````python
df_titanic_filtrado = df_titanic.dropna(subset="Age")

print(df_titanic_filtrado)
````

````python
primera_clase = df_sobrevivientes[df_sobrevivientes["Pclass"] == 1]

sobrevivientes_1ra_clase = primera_clase[primera_clase["Survived"] == 1]

dimensiones_1ra_clase_survived = sobrevivientes_1ra_clase.shape

porcentaje_1ra_clase = (dimensiones_1ra_clase_survived[0] * 100) / dimensiones[0]


segunda_clase = df_sobrevivientes[df_sobrevivientes["Pclass"] == 2]

sobrevivientes_2da_clase = segunda_clase[segunda_clase["Survived"] == 1]

dimensiones_2da_clase_survived = sobrevivientes_2da_clase.shape

porcentaje_2da_clase = (dimensiones_2da_clase_survived[0] * 100) / dimensiones[0]



tercera_clase = df_sobrevivientes[df_sobrevivientes["Pclass"] == 3]

sobrevivientes_3ra_clase = tercera_clase[tercera_clase["Survived"] == 1]

dimensiones_3ra_clase_survived = sobrevivientes_3ra_clase.shape

porcentaje_3ra_clase = (dimensiones_3ra_clase_survived[0] * 100) / dimensiones[0]


print(f"Porcentajes\n1ra Clase:   {porcentaje_1ra_clase} %\n2da Clase:  {porcentaje_2da_clase} %\n3ra Clase:  {porcentaje_3ra_clase} %")


````

````python
primera_clase_filtrado = df_titanic_filtrado[df_titanic_filtrado["Pclass"]== 1]
#Hombres 1ra clase
hombres_1ra_clase = primera_clase_filtrado[primera_clase_filtrado["Sex"]== "male"]

dimensiones_hombres_1ra_clase = hombres_1ra_clase.shape

suma_edades_hombres_1ra_clase = hombres_1ra_clase["Age"].sum()

promedio_edades_1ra_clase_hombres = suma_edades_hombres_1ra_clase / dimensiones_hombres_1ra_clase[0]
#Mujeres 1ra clase
mujeres_1ra_clase = primera_clase_filtrado[primera_clase_filtrado["Sex"]== "female"]

dimensiones_mujeres_1ra_clase = mujeres_1ra_clase.shape

suma_edades_mujeres_1ra_clase = mujeres_1ra_clase["Age"].sum()

promedio_edades_1ra_clase_mujeres = suma_edades_mujeres_1ra_clase / dimensiones_mujeres_1ra_clase[0]



segunda_clase_filtrado = df_titanic_filtrado[df_titanic_filtrado["Pclass"]== 2]
#Hombres 2da clase
hombres_2da_clase = segunda_clase_filtrado[segunda_clase_filtrado["Sex"]== "male"]

dimensiones_hombres_2da_clase = hombres_2da_clase.shape

suma_edades_hombres_2da_clase = hombres_2da_clase["Age"].sum()

promedio_edades_2da_clase_hombres = suma_edades_hombres_2da_clase / dimensiones_hombres_2da_clase[0]
#Mujeres 2da clase
mujeres_2da_clase = segunda_clase_filtrado[segunda_clase_filtrado["Sex"]== "female"]

dimensiones_mujeres_2da_clase = mujeres_2da_clase.shape

suma_edades_mujeres_2da_clase = mujeres_2da_clase["Age"].sum()

promedio_edades_2da_clase_mujeres = suma_edades_mujeres_2da_clase / dimensiones_mujeres_2da_clase[0]





tercera_clase_filtrado = df_titanic_filtrado[df_titanic_filtrado["Pclass"]== 3]
#Hombres 3ra clase
hombres_3ra_clase = tercera_clase_filtrado[tercera_clase_filtrado["Sex"]== "male"]

dimensiones_hombres_3ra_clase = hombres_3ra_clase.shape

suma_edades_hombres_3ra_clase = hombres_3ra_clase["Age"].sum()

promedio_edades_3ra_clase_hombres = suma_edades_hombres_3ra_clase / dimensiones_hombres_3ra_clase[0]
#Mujeres 3ra clase
mujeres_3ra_clase = tercera_clase_filtrado[tercera_clase_filtrado["Sex"]== "female"]

dimensiones_mujeres_3ra_clase = mujeres_3ra_clase.shape

suma_edades_mujeres_3ra_clase = mujeres_3ra_clase["Age"].sum()

promedio_edades_3ra_clase_mujeres = suma_edades_mujeres_3ra_clase / dimensiones_mujeres_3ra_clase[0]




print(f"Edad promedio hombres 1ra clase: {round(promedio_edades_1ra_clase_hombres,0)}\nEdad promedio mujeres 1ra clase: {round(promedio_edades_1ra_clase_mujeres,0)}\nEdad promedio hombres 2da clase: {round(promedio_edades_2da_clase_hombres,0)}\nEdad promedio mujeres 2da clase: {round(promedio_edades_2da_clase_mujeres,0)}\nEdad promedio hombres 3ra clase: {round(promedio_edades_3ra_clase_hombres,0)}\nEdad promedio mujeres 3ra clase: {round(promedio_edades_3ra_clase_mujeres,0)}")


````

````python
df_titanic_filtrado["MayorEdad"] = df_titanic_filtrado['Age'].apply(lambda x: 1 if x > 21 else 0)

print(df_titanic_filtrado)
````

````python
dimensiones_titanic_filtrado = df_titanic_filtrado.shape

df_sobrevivientes = df_titanic_filtrado[df_titanic_filtrado["Survived"] == 1]

df_sobrevivientes_menores = df_sobrevivientes[df_sobrevivientes["MayorEdad"] == 0]

dimensiones_sobrevientes_menores = df_sobrevivientes_menores.shape

porcentaje_sobrevivientes_menores = (dimensiones_sobrevientes_menores[0] * 100) / dimensiones_titanic_filtrado[0]

df_sobrevivientes_mayores = df_sobrevivientes[df_sobrevivientes["MayorEdad"] == 1]

dimensiones_sobrevientes_mayores = df_sobrevivientes_mayores.shape

porcentaje_sobrevivientes_mayores = (dimensiones_sobrevientes_mayores[0] * 100) / dimensiones_titanic_filtrado[0]


print(f"\nPorcentaje de sobrevientes menores de edad: {round(porcentaje_sobrevivientes_menores,0)} %\nPorcentaje de sobrevientes mayores de edad: {round(porcentaje_sobrevivientes_mayores,0)} %\n")
````
